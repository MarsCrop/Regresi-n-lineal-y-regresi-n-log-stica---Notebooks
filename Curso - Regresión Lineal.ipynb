# Machine Learning Ideas

### Generalización y regresión lineal

Antes de empezar el tema principal es necesario entender algunos conceptos básicos de machine learning para poder entender con más precisión cómo funciona la regresión. Uno de los conceptos importantes de los cuales se basa el método de regresión es el concepto de **generalización**. Generalizar en un contexto de uso de un algoritmo machine learning implica encontrar las instancias correctas de las **variables categóricas**. Las variables categóricas son simplemente conjuntos enumerables que pueden servir para denominar a otras instancias (datos) que se encuentran en un mismo plano, espacio o conjunto de datos.

### Un ejemplo crucial de generalización: analizando las futuras compras

Supongamos que un empresario desea medir el progreso de su empresa en términos de ventas alcanzadas respecto a las compras realizadas para mantenerla en funcionamiento y para lograr medir el progreso utiliza dos tipos de información, el primer tipo de información será un historial de unidades de mercancías compradas (llamadas en su caso, compras) El segundo tipo de información del cual dependerán, a su entender, las compras de mercancía para la empresa son las ventas. Desde una perspectiva matemática cada número de compras y ventas diarias se puede ubicar en una nube de puntos siempre y cuando posicionemos las unidades de compras hechas por día en el eje **x** y las unidades de ventas hechas por día en el eje **y**.

### Aplicando métodos científicos para comprender la magnitud del análisis

**Utilizaremos el método científico, en particular el método inductivo y el método deductivo**. ¿Qué tienen que ver los conceptos de machine learning con estos métodos? La **generalización** es una forma de aplicar estadísticamente el método inductivo. Dicho método consiste en validar aquellas hipótesis o teorías cuyas observaciones son correctamente atribuidas a las variables correspondientes de acuerdo a una denominación global hecha sobre las observaciones (datos). En términos estadísticos esto significa hallar recortes óptimos (es decir, que no den lugar a **contradicciones**) que permiten dividir una nube de puntos en grupos correspondientes a una expresión categórica (es decir basada en cierta distinción hecha sobre unos datos respecto de otros datos).
Las variables categóricas serán simplemente las denominaciones realizadas de forma enumerada (es decir, deben corresponderse a un orden categórico que quiero establecer, por ejemplo: no sirve de nada realizar una inducción entre homosexuales y personas cis, pero si tiene sentido realizar una inducción para hallar personas homofóbicas y personas a favor de la libertad sexual, las cuales podemos utilizar de forma enumerada como 0 y 1 respectivamente).
Si bien, pueden haber dos o más variables categóricas, **las contradicciones entre unas y otras variables se resuelven de manera dicotómica**. ¿Por qué? Porque **sin una resolución dicotómica los problemas de distinción entre las categorías se reducen de forma que no se termina de encontrar observaciones más concretas que permiten establecer la diferencia entre una categoría y todas las demás**. Ilustremos esto con un ejemplo mas.

### Reconociendo variables categóricas al generalizar

Supongamos que tengo información almacenada en una hoja de cálculo acerca de miles y miles de discos y quiero saber a qué género pertenecen todos los discos. La tarea de reconocer a qué género pertenece cada disco implica resolver el problema de reconocimiento entre discos que se presupone que pertenecen a un género y los discos que se presupone que pertenecen a otro género. Esto a su vez implica permutar los géneros de forma binaria para obtener diferentes subproblemas de los cuales luego se extraerán las **variables aplicables** al reconocimiento de un género **que aparecen en todos los subproblemas**. Los subproblemas en este caso serían rock vs blues, rock vs jazz, blues vs jazz.
En términos matemáticos, la mejor resolución de cada subproblema dada una nube de puntos -de la cual ya encontramos una tendencia o ajuste- nos permitirá categorizar sobre nueva información. De la misma forma que logramos encontrar **ajustes entre variables en x e y, podremos encontrar ajustes para nuevas variables x y nuevas variables y**. Supongamos que queremos reconocer mediante información fotográfica perros y gatos. Luego de realizar los ajustes estadísticos correspondientes, la generalización hecha sobre las fotos utilizadas para reconocer perros y gatos será mucho mas aceptable si también permitiese reconocer fotos de cachorros y de gatos bebés como perros y gatos respectivamente.
A esto nos referimos cuando hablamos de **generalización**. No podremos decir que todos los discos en la nube de puntos son de rock, pero si establecer recortes en dicha nube y establecer márgenes de separación que permitan establecer que **todos los puntos en el subconjunto A pertenecen a la variable categórica A y no a los puntos en el subconjunto B que pertenecen a la variable categórica B**.
Dicho esto, al problema de descartar una o más variables que nos permitan hallar subconjuntos (correspondientes a variables categóricas) y al de categorizar en base a observaciones que no tienen relación con su categoría asignada, aparece un nuevo concepto que, en términos prácticos, nos permite realizarnos las autocríticas necesarias a la hora de entrenar un algoritmo de regresión.

### Regresión unidimensional

Se denomina **regresión lineal** a un modelo utilizado para poder medir las relaciones de información en estructuras de datos. 
Siendo **x** una columna de datos en la estructura correspondiente a un rasgo, item, o característica y siendo **y** una fila de datos correspondiente a otro item o característica, se deben obtener valores para **y (el cual representa los valores de ajuste/tendencia del conjunto de datos)** resolviendo la ecuación **w * x + b** en donde:

* **w (width, ancho)** mide la influencia/distancia de los datos en x sobre la linea de regresión **Y**
* **b (intercepto, constante)** es una constante que puede servir a redireccionar los valores en w veces x hacia un punto en el plano donde **los valores contengan menos información residual**.

Los valores de **w** y de **b** se calculan utilizando el **método de mínimos cuadrados**, el cual optimiza el ajuste/tendencia **minimizando la suma de residuos cuadráticos**. La diferencia entre utilizar el método de mínimos cuadrados y utilizar la regresión lineal es que en la regresión lineal **la ecuación se puede expresar como una forma lineal, mientras que el método de mínimos cuadrados solo busca encontrar un ajuste óptimo para los valores en la ecuación mejorando el valor de w dada una constante b**. Además, **el método de mínimos cuadrados solo resuelve la ecuación para (w veces x) + b, mientras que la regresión lineal busca el óptimo ajuste descomponiendo las variables optimizadas para predecir el devenir de mas variables tomando en cuenta que el ajuste es óptimo, ya que la varianza se ha reducido en menor medida**. 

Las utilidades de la regresión lineal suelen ser:

* hallar una tendencia entre la cantidad de los datos que permite establecer un límite entre datos que forman parte de un entorno positivo y otros que forman parte de un entorno negativo.
* predecir valores futuros de los datos que estamos obteniendo. Por ejemplo: si yo tengo una columna de datos x de tamaños de casas y una columna de datos y de precios de casas, luego de encontrar valores de w y de b apropiados puedo predecir el precio futuro de una casa en el conjunto de datos x.


Al utilizar scikit-learn contaremos con herramientas básicas y simples de utilizar para resolver problemas de regresión. Nuestro primer programa de regresión lineal resolverá problemas unidimensionales, donde contamos con una matrix **x** (de una columna y varias filas) y otra matriz **y** (que consta de una fila y varias columnas) para obtener **un sólo valor numérico para w y otro para b**.
Utilizaremos scikit-learn para programar los procesos de forma muy simple, pero antes, comprenderemos cómo se utiliza el método de mínimos cuadrados.

### Regularización y regresión lineal

**Regularizar en términos matemáticos implica agregar una nueva variable a una función para prevenir el sobreajuste, es decir, prevenir las categorizaciones hechas de manera errónea. Utilizando una función de pérdida (loss también entendida como función de error) para medir el progreso de optimización, la regularización limita un poco más la búsqueda de un ajuste óptimo al utilizar la convergencia de probabilidad entre el valor óptimo de w y la variable de costo (C). Es decir, la regularización nos permite hallar más rápido el límite de convergencia en el cual queda establecido que el valor de w es probable de ser óptimo respecto al límite de coste establecido**.

###### ¿Qué tiene que ver esto con la regresión lineal? 

Bueno, si tenemos que encontrar una solución óptima no hay que dar lugar a ajustes inapropiados, que estan muy alejados del valor promedio del total de la información respecto al ancho **w**. Si no hubiese regularización, el límite encontrado le dará mayor peso a datos que son irrelevantes para encontrar el ajuste y la ecuación **w veces x + b** nos daría como resultado una pendiente o recta que se va a ir corriendo no sólo en una dirección inadecuada, sino que estará ubicada en un punto donde el ajuste no será muy bueno.

### Aplicando la regularización para regresión lineal. Usos y dificultades

En este ejemplo utilizaremos una forma regularizada de regresión lineal donde utilizamos una matriz de identidad de las mismas dimensiones que x (esto es, una matriz completada con ceros pero cuya diagonal esta hecha en base a unos) como intercepto inicial, es decir, utilizamos el resultado de **(x veces x transpuesta) + la identidad de x** como punto de partida para encontrar el coeficiente que minimiza la suma de residuos (**w**) y luego obtenemos el intercepto correcto utilizando los valores promedios de x e y respectivamente, siendo x multiplicado por el coeficiente **w**.
El método aplicado en este ejemplo no es malo, pero no lo podemos ajustar, ya que la optimización mediante mínimos cuadrados no permite utilizar un parámetro de regularización como si podemos a la hora de aplicar descenso de gradiente. La regularización mediante mínimos cuadráticos esta determinada por la matriz de identidad y por los conjuntos de datos en **x**, por lo que el resultado de ajuste obtenido es **determinista, o sea, la solución obtenida por el ajuste es definitivo y, por lo tanto es mucho más difícil de validar en otros contextos de análisis (por ejemplo, los contextos de análisis donde dividimos los datos para probar el modelo de ajuste con datos que no fueron entrenados)**.

###### Respecto al método de regularización con mínimos cuadrados, los pasos a seguir son los siguientes:
    * utilizamos nuestras variables en x e y como es usual
    * utilizamos el método least squares de numpy utilizando como variable en x la nueva expresión de x basada en la resolución de (w veces x) + b, la cual es (x veces x transpuesta) + la identidad de x.
    * la variable en y es simplemente el producto interior K(x,x.T) de nuestra variable x respecto a su forma transpuesta
    * despues de utilizar least squares para obtener nuestro valor de w, calculamos el valor de b. Por último obtenemos la linea de regresión resolviendo para nuestros datos x la ecuación (w veces x) + b

```python
import numpy as np
from numpy.linalg import lstsq

x = np.mat(np.random.randint(1,16, 20)).T
y = np.mat(np.random.randint(2,20, 20))

#utilizamos least squares tomando como argumentos los productos interiores de x en relación a su matriz transpuesta y con un 'intercepto' basado en la diagonal de x
w = lstsq(x @ np.mat(x).T +  np.identity(x.shape[1]), x @ np.mat(y))[0]

b = np.average(y) - (w * np.average(x))

#liney resulta de calcular la media resultante del cálculo hecho previamente, encontrando así un resultado con la menor varianza posible
liney = np.mean((w * x + b).T, axis=0)

plt.plot(x, liney.T)

plt.scatter(np.array(x.T),np.array(y))
plt.show()
```

### Descenso de gradiente

Debido al subproblema que agrega la regularización, algunas personas prefieren utilizar otros algoritmos de optimización en lugar de utilizar mínimos cuadrados. Un algoritmo muy popular de resolución del problema es el **descenso de gradiente**. El descenso de gradiente consiste en utilizar las variables en **x** e **y** para encontrar un **óptimo local basado en un mínimo local**. **Un mínimo local es un punto en la nube de puntos en donde las observaciones obtenidas son necesariamente óptimas y donde podemos utilizar variables que, dentro del radio en que se encuentre el mínimo local, son las que nos permiten obtener la solución para una función que requiere de la búsqueda de un resultado mínimo/minimizable**. **El óptimo local es aquel mínimo local encontrado que contiene mayor probabilidad de contener varios datos alrededor como para obtener una solución óptima a la búsqueda de w y b**.
**El método de descenso de gradiente consiste en encontrar un óptimo local (coordenada óptima en la nube de puntos) para hallar valores de w y de b tomando como punto de partida un dato en x para ir en dirección descendiente hasta encontrar un gradiente (campo de vector direccionado) donde haya convergencia luego de obtener los valores de w y b dentro del campo de datos. Utilizando este método vamos considerando solamente los datos mas relevantes a medida que el campo vectorial disminuye y nos ubicamos en la dirección correcta para solucional el problema de optimización de búsqueda de variables**.
En comparación respecto a los mínimos cuadrados, el método de descenso de gradiente requiere **utilizar x como resultado de un producto interior entre todos los datos en x bajo la expresión K(x,x.T), siendo esta expresión simplemente una matriz Gram que resulta de multiplicar todos los datos en la matriz de x unos contra otros para poder utilizar una matriz que utilize mejor las conexiones entre todos los puntos en x a la hora de encontrar un ajuste en y mediante redireccionamiento de los datos en x hacia el óptimo local**.
Nuetra variable de regularización/tolerancia es el **tamaño de aprendizaje, un factor que aumenta o disminuye el campo de búsqueda a la hora de optimizar nuestra solución. Usualmente el tamaño de aprendizaje es bajo para no tomar direccionamientos demasiado largos ni desviar nuestra solución óptima hacia otro lado**.

### Ejemplo de descenso de gradiente

Los pasos en este ejemplo son los siguientes:
    * Importamos numpy con el alias np
    * Importamos pyplot para visualizar los resultados
    * Definimos una función SGD para entrenar los datos de regresión linear utilizando el método de descenso de gradiente estocástico. El método estocástico a diferencia del método normal de descenso de gradiente, simplemente minimiza la búsqueda del óptimo local siempre y cuando el resultado de direccionar los valores en K(x, x.T) como producto de la relación con sus variables en y nos de un producto menor a 1 (variable que representa un paso hecho hacia abajo). Mientras el resultado de direccionarnos hacia abajo sea menor a 1, podremos tomar el vector que se halla en el gradiente como relevante para hallar el intercepto b.
    Los argumentos utilizados para la función son los valores en x e y en conjunto con el tamaño de aprendizaje **lr**.
    * La linea de regresión en y resultará de resolver nuestra ecuación de regresión (w veces x) + b. Finalmente visualizamos la nube de puntos y la linea de regresion.

```python
import numpy as np
import matplotlib.pyplot as plt

def SGD(x, y, lr):
    a = np.zeros(len(x))
    x = x @ x.T # aquí @ es un operador para obtención de producto interior dadas matrices de dimensiones (m,n) y (n,m)
    for i in range(20): #20 es la cantidad de iteraciones que utilizaremos para correr el algoritmo y encontrar convergencia
        iterations = 1
        for tau in range(len(a)): #iteramos sobre todos los x_tau hasta computar cada valor de alfa (a) en cada iteración
                wx = a @ x[:,tau] #el ancho entre una gradiente y la próxima se calcula multiplicando alfa por x_tau
                a[tau] *= (1 - 1/iterations)
                if(y[tau]*wx < 1): #verificamos que el desplazamiento del vector no sea tan grande
                    a[tau] += y[tau]/(lr * iterations)
                iterations += 1
        b = (-(((x[a>0] * wx.T).max() + (x[a>0] * wx.T).min())/2)) #computamos los valores del intercepto luego de cada iteración del algoritmo               
    return wx, b

x = np.array([[ 0.],[ 1.],[ 2.],[ 3.],[ 4.]])
y = np.array([2, 2, 3, 3, 4])

wx, b = SGD(x,y,0.002)
print(wx * x + b) #valores de la línea de regresión
plt.plot(x, wx*x+b)
plt.scatter(x,y)
plt.title('Descenso de gradiente con regularización/tamaño de aprendizaje 0.002')
plt.show()

wx, b = SGD(x,y,0.5)
plt.plot(x, wx*x+b) #valores de la línea de regresión
plt.scatter(x,y)
plt.title('Descenso de gradiente con regularización/tamaño de aprendizaje 0.5')
plt.show()

wx, b = SGD(x,y,0.05)
plt.plot(x, wx*x+b) #valores de la línea de regresión
plt.scatter(x,y)
plt.title('Descenso de gradiente con regularización/tamaño de aprendizaje 0.05')
plt.show()
```

   Con estas lineas ya utilizamos el método fit para obtener las variables de **w** y de **b**.
   El sistema devolverá un error si la cantidad de datos en **x** no es la misma que la de **y**, porque no se podrá optimizar sobre una variable con mas valores que otra, porque **un dato por si solo no nos permitirá marcar tendencia, sino que es la relación entre el dato de esa variable y el dato de la otra variable (sean x o y) las que nos permiten calcular la tendencia**.

# Regresión multidimensional utilizando scikit-learn

### Regresión multidimensional

Ya hemos visto en la etapa anterior del curso varias formas de realizar regresión lineal unidimensional tomando en cuenta ventajas y desventajas de utilizar uno u otro algoritmo. Ahora nos concentraremos en aplicar el método de regresión multidimensional utilizando solamente scikit-learn y nuestra comprensión de la multidimensionalidad de los datos (simplemente tener en cuenta que una matriz de datos puede contener más de un dato en la misma fila o columna). Utilizando scikit-learn podemos realizar regresión lineal multidimensional utilizando varias columnas de x y varias columnas de y. Este programa de regresión lineal predecirá la tendencia de dispersión del ébola en África de manera multidimensional, contando con una **matriz x de varias columnas de datos** y otra **matriz y de varias columnas**, generando **coeficientes de w y de b expresados n veces, respecto a la cantidad de columnas en x e y**. 

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import MinMaxScaler

import pandas as pd

#importamos la base de datos del ébola en .csv
df = np.array(pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_ebola.csv'))

x1 = np.nan_to_num(df[:,5].tolist())

#transformamos las columnas de datos
x = MinMaxScaler().fit_transform(np.hstack([np.vstack((df[:,1], x1)).T,np.vstack((df[:,3], df[:,4])).T]))

X, Y = x[:,(0,1)],x[:,(2,3)]
```

En primer lugar, tuvimos que escalar los valores de los datos en la base de datos para poder encontrar un ajuste más preciso. Para prevenir generalizaciones erróneas correremos el método de regresión lineal de scikit learn de forma iterativa para obtener una línea de regresión por cada columna de datos.
    
### Predecir la regresión multidimensional en y

Llamaremos **predecir** a realizar el cálculo final de la ecuación **w * x + b** que nos devolverá **más de una linea de regresión** (se genera por cada columna de datos una linea de regresión nueva).

```python
import matplotlib.pyplot as plt

ys = []
for i in range(X.shape[1]):
    twoD_lr = LinearRegression()
    twoD_lr.fit(np.mat(X[:,i]).T,Y[:,i].T)
    line_y = twoD_lr.predict(np.mat(X[:,i]).T) #predecimos los resultados sobre y
    ys.append(line_y)

    plt.plot(np.mat(X[:,i]).T, np.array([line_y]).T)

plt.show()
```

Aunque los valores no sean en orden ascendiente, ¡hemos obtenido dos líneas rectas!. Asi funciona la regresión multidimensional, se deben encontrar tantas soluciones/ajustes como columnas de datos existentes en una estructura de datos.

# Regresión polinomial utilizando scikit-learn
   
### Regresión Polinomial

Cuando utilizamos valores de x y de y expresados de forma polinómica, estamos frente a un caso de **regresión no lineal** dado que la transformación de la forma polinómica de x no será directamente lineal, generando una pérdida en el calculo residual (la varianza será mayor que la esperada si se intentan resolver regresiones lineales con formas polinómicas sin ningun criterio establecido).
Este inconveniente a la hora de resolver problemas de regresión lineal se suele conocer como la **maldición de la dimensionalidad**. La maldición de la dimensionalidad reconoce que la cantidad de datos aumenta a medida que agregamos nuevas columnas a las matrices, y de hecho, con un poco intuición podemos darnos cuenta que al generar nuevas columnas que representan a x potenciada a cada enésima columna el espacio ocupado en la nube de puntos será muchísimo mayor al de una regresión multidimensional. Por ende, **cuando la matriz que utilizamos es aun mas dispersa se vuelve difícil encontrar un ajuste perfecto**. 

### ¿Cómo romperemos el hechizo?

Para romper con la maldición de la dimensionalidad simplemente utilizaremos una matriz de datos en la cual se incluyen columnas de valores potenciados. En este caso la cantidad de datos aparece unidimensionada, por lo que crearemos nuevas columnas de x potenciadas n veces de forma gradual. Vale aclarar que los valores del vector a utilizar para la regresión lineal deberían oscilar entre la primer potencia cuadrada del segundo valor para que el ajuste sea exacto y no se desvíe del dominio correspondiente.
Resolver la regresión polinomial nos dará ciertas ventajas:
* Utilizaremos formas derivadas de w * x para el ajuste final, por lo que la linea de regresión se correrá hacia ciertos datos y en otros el ajuste será recto.
* Podemos resolver problemas de regresión en los que la matriz original de x es muy densa. Simplemente reducimos el ruido en la información generando formas polinómicas para que el valor promedio de los datos nos de ventajas para dar con un buen ajuste.

### En scikit-learn generar columnas de datos de forma polinómica no cuesta mas que unas líneas
    * importamos el método LinearRegression de scikit-lean
    * importamos el método PolynomialFeatures, para entrenar la regresión lineal sobre datos expresados de forma polinómica. Esto es equivalente a utilizar una función que intente encontrar las relaciones entre x e y de forma gradual
    * importamos la base de datos del ébola para encontrar una tendencia polinómicamente, haciendo esto podremos discriminar mejor acerca de dónde habrá un mayor brote de ébola.
    * iniciamos el método PolynomialFeatures con un grado de 3 (los rasgos polinómicos suelen ser de hasta dos grados para no dimensionar mucha información)
    * usamos el método fit_transform de Polynomial Features, el cual solamente generará nuevas columnas en x las cuales serán solamente los datos en x elevados a la potencia correspondiente a la enésima columna
    * usamos el mismo método con nuevas variables para construir rasgos polinómicos de coordenadas en y

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn.preprocessing import MinMaxScaler

import pandas as pd

#utilizamos pandas para leer el conjunto de datos en .csv
df = np.array(pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_ebola.csv'))

#reemplazamos las celdas vacías en el archivo .csv por 0s
x1 = np.nan_to_num(df[:,5].tolist())

x = MinMaxScaler().fit_transform(np.vstack((df[:,1], df[:,4])).T)

X, Y = x[:,0],x[:,1]

#inicializamos el método PolynomialFeatures
poly = PolynomialFeatures(degree=3, include_bias = False)

#reconstruimos la matriz de datos en x
X_ = poly.fit_transform(np.mat(X))

#transformamos la matriz de datos en y
predict_ = poly.fit_transform(np.mat(Y))
```

El parámetro *degree* del método PolynomialFeatures es muy importante ya que indicará cuántas veces vamos a reconstruir a x. En la mayoría de los buenos análisis el grado de la función no suele ser mayor que 2. Luego podemos realizar una regresión lineal sobre cualquier fila de datos nuestra matriz de datos en y que resulta también de la conversión en forma polinómica de los datos.

```python
polyr = LinearRegression()
polyr.fit(X_,predict_)
```


¡El conjunto de datos en x y el conjunto de datos en y han sido validados y los valores de w y de b han sido calculados!.

### Sobreajuste (Overfitting)

Ahora, ¿qué ocurre cuando nuestra línea de regresión aparece en un punto no intermedio en el plano? Eso será indicio de un concepto se conoce como **overfitting (sobreajuste)**. Se conoce como **sobreajuste a realizar un ajuste excesivo sobre nuestros datos de forma que las tendencias obtenidas no permitirán encontrar subconjuntos adecuados, por lo que habrán datos muy mal categorizados, teniendo asi errores altos en el ajuste**. Los sobreajustes se encuentran cuando la optimización de nuestro **weight o factor de importancia hecho sobre los datos** no es óptimo, por lo que el error en el ajuste es alto. Esto se suele deber a que no utilizamos los límites (parámetros establecidos por nosotros los humanos) que permiten reducir el error adecuado.
**El sobreajuste esta relacionado a cómo realizamos lo que se conoce como relajación lagrangiana, es decir, resolver el problema de ajuste utilizando variables extra que utilizamos para substraer información de acuerdo a su relevancia, reduciendo los problemas de optimización a valores mejor minimizados o maximizados**. Si nuestra **variable slack, o valor independiente que aparece como penalizador es muy alto, el ajuste estará maximizado en exceso y si nuestro valor independiente penalizador es muy bajo el ajuste estará minimizado en exceso y no hallaremos un ajuste apropiado que nos permita establecer tendencias incluso con nuevas variables**.
Volvamos al primer ejemplo, el de el **análisis comercial**. ¿Cómo podría el empresario aplicar estos conceptos? Siguiendo estos pasos:

Teniendo este objetivo en mente sus variables categóricas serán aquellas referentes a la caída o incremento de ventas, las cuales en términos estadísticos se tomarán como 0 y 1 respectivamente.

##### Una vez que ya estan claras las variables categóricas podremos comprender el ajuste de una forma analítica. ### Midiendo el error en la regresión

A continuación, siguiendo con el ejemplo de análisis de compras y ventas, vamos a medir el error en las regresiones que hagamos. Medir el error utilizando scikit-learn requiere de utilizar dos funciones básicas. La que mide el **coeficiente de determinación** y la que mide el **error medio cuadrático**.
El error se mide utilizando las variables de **y** y las variables de ajuste generadas con el método **predict** de **LinearRegression**.
El **coeficiente de determinación** se utiliza para medir el error de acuerdo a la información residual y la información de regresión. 
El **error cuadrático medio** es simplemente una expresión al cuadrado del **error cuadrático**, el cual mide la diferencia entre las variables de **y** predecidas y las variables de **y** utilizadas para un ajuste. Cuando la diferencia disminuya hacia 0, el error será menor. La diferencia entre utilizar solamente el error cuadrático y potenciarlo es que si no potenciamos el error habrá nueva información que podrá incrementar o disminuir el error sin tomar en cuenta ningun valor promedio durante la optimización del problema. Al potenciar el error cuadrático es más facil encontrar un valor promedio que me permita medir la eficacia del algoritmo y la robustez en el análisis.
Esto significa que el error será pequeño si la línea de regresión se encuentra ubicada entre información mas importante. Nuestra medida de interés debe ser el error medio cuadrático, ya que al estar basado en la búsqueda de buena información entre todas las muestras, nos podrá decir si la regresión es de veras exacta si agregamos nueva información.
A continuación escribiremos una pequeña función que nos permite medir el coeficiente de determinación entre estas dos variables.

### Midiendo el error de regresión con el coeficiente de determinación

    * primero importamos numpy
    * utilizaremos la línea de ajuste liney como el valor de ajuste predecido por la regresión y las cordenadas en y correspondientes con las ventas para comparar los resultados entre el ajuste hecho y los valores reales en y
    * definimos una función llamada coeficiente de determinación la cual obtiene un promedio de los valores en y la cual se substrae de las muestras en y para establecer cuál es la información importante (en promedio) en y respecto a la línea de regresion. Una vez hecho esto calculamos el resto de la división entre la información relevante para la regresión y la información que sobra

```python
#r2 error
def coefficient_of_determination(y, liney):
    ybar = np.sum(y)/len(y)        
    ssreg = np.sum((liney-ybar)**2) 
    sstot = np.sum((y - ybar)**2)
    return (ssreg / sstot) - 1

print(coefficient_of_determination(ventas,liney))
```

El error basado en coeficiente de determinación nos da como resultado una variable negativa. Esto se debe a que la relacion entre la suma de  regresión y la suma de la información residual/insuficiente es casi nula, lo que nos da a entender que no hay mucha  información residual durante el análisis (la búsqueda de los valores w y b). Además, la distancia entre la regresión y buena parte de los datos es mucha y solamente se toman en cuenta los valores cercanos a la línea de regresión.
Veamos que ocurre cuando medimos el error cuadrático medio de estos valores.

### Utilizando el error cuadrático medio para mensurar el error de regresión
    * importamos numpy y el método mean squared error de scikit-learn
    * gracias a scikit-learn podemos obtener el error cuadrático medio sin programar nada, solamente hay que llamar a la función correspondiente

```python
from sklearn.metrics import mean_squared_error  

print(mean_squared_error(ventas,liney))
```

##### El error pasó de -0.9 a 17 puntos. ¿A qué se debe tanta diferencia a la hora de cambiar la forma en que medimos el error? Esto se debe a que los datos no estan escalados apropiadamente, por lo que la información proporcionada no dará un valor promedio que oscile entre un mínimo y máximo de valores bajos. Cuanto más pequeña sea la escala de los valores más fácil es encontrar el error utilizando machine learning porque los valores se pueden promediar a una escala que encuentra un promedio más ajustado. Veamos como podemos corregir este problema

    * Escalaremos los datos entre 0 y 1 importando el algoritmo MinMaxScaler de scikit-learn y volveremos a aplicar el método de regresión lineal sobre los datos ya escalados
    * Mediremos el error para asegurarnos que el ajuste mantiene relevancia

Mientras la diferencia entre la variable a predecir y la predecida sea 0 el error será menor. Es decir, el error cuadrático medio será menor mientras los valores de la línea de regresión oscilen entre los valores de la variable **y** para predecir.


```python
import numpy as np
import matplotlib.pyplot as plt 

#generación de datos de forma aleatoria
ventas = np.random.randint(1,16, 20)
compras = np.random.randint(2,20, 20)
labels = np.random.randint(0,2,20)

from sklearn.preprocessing import MinMaxScaler

#escalamos los datos generados para que oscilen entre 0 y 1
ventas = MinMaxScaler().fit_transform(np.mat(ventas).T)
compras = MinMaxScaler().fit_transform(np.mat(compras).T)

from sklearn.linear_model import LinearRegression

#regresión lineal
lr = LinearRegression()

#ajuste
lr.fit(compras, ventas)

#predicción
liney = lr.predict(compras)

#variables categóricas
targets = []
for i in range(len(ventas)):
    if ventas[i] > liney[i]:
        targets.append([1])
    else:
        targets.append([0])

        #visualización
plt.scatter(compras, ventas, c = targets)
plt.plot(compras, liney)
plt.title('Disminuyendo el sobreajuste en análisis de compra y ventas')
plt.xlabel('Compras')
plt.ylabel('Ventas')
plt.show()

def coefficient_of_determination(y, liney):
    ybar = np.sum(y)/len(y) #promediamos los valores en y       
    ssreg = np.sum((liney-ybar)**2) #sumamos la diferencia entre la línea de regresión con el promedio en y
    sstot = np.sum((y - ybar)**2) #sumamos la diferencia entre los datos en y con el promedio en y
    return (ssreg / sstot) - 1

print('Coeficiente de determinación: ' + str(coefficient_of_determination(ventas,liney)))

from sklearn.metrics import mean_squared_error  

print('Error cuadrático medio: ' + str(mean_squared_error(ventas,liney)))
```


### ¿Cómo puede un empresario que realiza estos análisis utilizar las medidas de error de una forma más práctica?

Es necesario realizar **validación cruzada**, esto es, dividir la cantidad de muestras en un porcentaje de datos para el cálculo de regresión/ajuste y otro porcentaje de datos para probar la calidad del ajuste utilizando información que no fue optimizada. Si el error es bajo en el uso de las muestras de prueba, quiere decir que el modelo de regresión ya cuenta con valores apropiados de **w** y de **b**.

```python
#utilizamos los primeros siete datos como datos de entrenamiento/ajuste
x_train = compras[:10] 
y_train = ventas[:10]

#utilizamos los tres restantes como datos de prueba
x_test = compras[10:]
y_test = ventas[10:]
```

Para resolver el análisis de compras y ventas con validación cruzada utilizaremos el 50% de los datos para calcular los valores de w y b y el resto lo utilizaremos para validar el ajuste.
Luego utilizaremos **fit** sobre los datos de ajuste y **predict** sobre los datos de ajuste y de prueba respectivamente para medir los errores. **Si el error de prueba es mas pequeño que el error de entrenamiento, el modelo puede ser validado. Además, si el error de ajuste/entrenamiento es bajo y el de prueba es mas bajo todavía, estamos realizando un análisis correcto**. Vamos a ver qué ocurre.

```python
oneD_lr = LinearRegression(n_jobs=-1) 

#ajuste real/de entrenamiento
oneD_lr.fit(x_train,y_train)

#predicción real/de entrenamiento
yhat_train = oneD_lr.predict(y_train)

#predicción de prueba
yhat_test = oneD_lr.predict(y_test)
```

Ahora mediremos el error de análisis con mean squared error y con el coeficiente de determinación.

```python
print("Error cuadrático medio de entrenamiento: " + str(mean_squared_error(y_train,yhat_train)))       
print("Error cuadrático medio de prueba: " + str(mean_squared_error(y_test,yhat_test)))
print("R2 error de entrenamiento: " + str(coefficient_of_determination(y_train,yhat_train)))
print("R2 error de prueba: " + str(coefficient_of_determination(y_test,yhat_test)))
```

Como verán tenemos mejores resultados utilizando el error cuadrático medio, **el error de entrenamiento es menor al error de prueba, por lo tanto podemos utilizar los valores de w y b en mas datos**.
